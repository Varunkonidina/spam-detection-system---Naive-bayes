
GET DUMMY'S FOR SPAM AND ADD IT TO THE DATA FRAME 

```python
dummy=pd.get_dummies(df.v1)
df["spam"]=dummy["spam"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.v2,df.spam,test_size=0.2)
```

# BUILD MODEL

USE PIPELINE TO CONVERT TEXT INTO NUMETRIC DATA AND TRAIN THE MULTINOMIAL MODEL

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
model=Pipeline([
    ("vect",CountVectorizer()),
    ("bayes",MultinomialNB())
])
```

```python
model.fit(x_train,y_train)
```

# TEST THE MODEL

```python
model.score(x_test,y_test)
```
